```python
import torch
import torch.nn as nn
import torchvision
```

```python
model = torchvision.models.resnet18()
model.conv1 = nn.Conv2d(6, 64, kernel_size=(7, 7), stride=(2, 2), padding=(3, 3), bias=False)
model.fc = nn.Sequential(nn.Linear(in_features=512, out_features=1, bias=True), nn.Sigmoid())
print(model)
```
